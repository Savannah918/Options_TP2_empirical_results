```python
# --------- store all the forward price from 2020 to 2023 feb to a dictionary, 
# --------- where key are (as_of_date, exdate), while value being the forward_price

import pandas as pd

def forward_price_dict(file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Create a dictionary with (column1, column3) as key and column4 as value
    forward_dict = dict(zip(zip(df.iloc[:, 1], df.iloc[:, 2],df.iloc[:, 3]), df.iloc[:, 4]))
    
    with open('2020_2023_forward_price_dict.pkl', 'wb') as f:
        pickle.dump(forward_dict, f)
        
    return forward_dict


result = forward_price_dict('./Forward_price 2020-2023feb.csv')
print(len(result))
```
